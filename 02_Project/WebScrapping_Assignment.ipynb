```python
from bs4 import BeautifulSoup
import requests
```

```python
# Write a python program to display all the header tags from wikipedia.org.

page = requests.get('https://en.wikipedia.org/wiki/Main_Page') # Send request to webserver to get the source code

soup = BeautifulSoup(page.content) # To Assign the page content to the variable soup

#titles = soup.find_all(["h1", "h2", "h3", "h4", "h5", "h6"])
#print('List all the header tags :', *titles, sep='\n\n')

header_tags = ["h1", "h2", "h3", "h4", "h5", "h6"]
for tags in soup.find_all(header_tags):
    print(tags.name + ' -> ' + tags.text.strip())
```

```python
# Write a python program to display IMDB's Top rated 100 movies' data (i.e. name, rating, year of release) and make data frame.

page = requests.get('https://www.imdb.com/chart/top/?ref_=nv_mv_250') # Send request to webserver to get the source code

soup = BeautifulSoup(page.content) # To Assign the page content to the variable soup
movietitle = soup.find_all('td', class_="titleColumn")
movietitle

movies = []

for movie in movietitle:
    movie = movie.get_text().replace('\n', "")
    movie = movie.strip(" ")
    movies.append(movie)
movies[:100]
    
scraped_ratings = soup.find_all('td', class_='ratingColumn imdbRating')
ratings = []
for rating in scraped_ratings:
    rating = rating.get_text().replace('\n', '')
    ratings.append(rating)
ratings

import pandas as pd
data = pd.DataFrame()
data['Movie Names'] = movies[:100]
data['Ratings'] = ratings[:100]
data
```

```python
# Write a python program to display list of respected former presidents of India (i.e. Name, Term of Office) from https://presidentofindia.nic.in/former-presidents.htm

page = requests.get('https://presidentofindia.nic.in/former-presidents.htm') # Send request to webserver to get the source code

president_soup = BeautifulSoup(page.content) # To Assign the page content to the variable soup
tag = ["h3"]
presidentname = []
for tags in president_soup.find_all(tag):
    tags = tags.get_text().strip()
    presidentname.append(tags)
presidentname
    
tag2 = ["p"]
term = []
for tag1 in president_soup.find_all(tag2):
    tag1 = tag1.get_text().strip()
    tag1 = tag1.replace('Term of Office: ', "")
    term.append(tag1)
term.remove("http://pranabmukherjee.nic.in")
term.remove("http://pratibhapatil.nic.in")
term.remove("http://abdulkalam.nic.in")
term[:13]

import pandas as pd
presidentlist = pd.DataFrame()
presidentlist['President Name'] = presidentname
presidentlist['Term of Office'] = term[:13]
presidentlist
```
